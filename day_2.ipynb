```python
fruits = ["apple", "banana", "mango"]
print("Initial:", fruits)

# access
print(fruits[0], fruits[1], fruits[2])

# modify
fruits[1] = "orange"

# add
fruits.append("grape")

# remove
fruits.remove("apple")

# pop last
last = fruits.pop()

# sort
fruits.sort()

print("Final List:", fruits)
print("Popped Item:", last)
```

```python
nums = {1, 2, 2, 3, 4, 4, 5}
print("Unique set:", nums)

nums.add(6)
nums.discard(3)

print("Final set:", nums)

# membership check
print(4 in nums)
print(10 in nums)
nums.add(11)
print(nums)
```

```python
student = {
    "name": "Ajay",
    "age": 23,
    "university": "FAU"
}

print(student)

# access
print(student["name"])

# modify
student["age"] = 24

# add
student["country"] = "USA"

# remove
student.pop("university")

print("Updated dictionary:", student)
```
