```python
#Student Result Analyzer
students = [
    {"name": "Ajay", "marks": 85},
    {"name": "Rahul", "marks": 62},
    {"name": "John", "marks": 45}
]

for student in students:
    if student["marks"] >= 75:
        print(student["name"], "→ Distinction")
    elif student["marks"] >= 60:
        print(student["name"], "→ Pass")
    else:
        print(student["name"], "→ Fail")
```
